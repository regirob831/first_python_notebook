```python
merged = pandas.merge(pot, contribs, on="calaccess_committee_id")
```

```python
merged.groupby(["committee_position", "contributor_employer"]).amount.sum().reset_index().sort_values("contributor_employer")
```

```python
support.groupby(['contributor_lastname','contributor_firstname']).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_supporters = support.groupby(
    ["contributor_firstname", "contributor_lastname"]
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_supporters.head(10)
```

```python
top_supporters.amount.plot.bar()
```

```python
top_supporters.amount.plot.barh()
```

```python
chart = top_supporters.head(5).amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_lastname)
```

```python
top_supporters['contributor_fullname'] = top_supporters.contributor_firstname + " " + top_supporters.contributor_lastname
```

```python
top_supporters.head()
```

```python
chart = top_supporters.head(5).amount.plot.barh()
chart.set_yticklabels(top_supporters.contributor_lastname)
```

```python
top_supporters.to_csv("top-supporters.csv",index=False)
```

```python
top_oppose = oppose.groupby(
    ["contributor_firstname", "contributor_lastname"]
).amount.sum().reset_index().sort_values("amount", ascending=False).head(10)
```

```python
top_oppose['contributor_fullname'] = top_oppose.contributor_firstname + " " + top_oppose.contributor_lastname
```

```python
chart = top_oppose.head(5).amount.plot.barh()
chart.set_yticklabels(top_oppose.contributor_lastname)
```

```python
top_oppose.to_csv("top_oppose.csv",index=False)
```
